# 🔗 Step 2: Data Merging

Joining all CSV files into one main dataframe.

## Why are we merging?
- `results.csv` is the **base** of our project
- It contains IDs like `raceId`, `driverId`, `constructorId` — not actual names
- We need to join other files to get **actual information**

## Merge Order:
1. results + races → add year, circuit, round
2. + drivers → add driver name, nationality
3. + constructors → add team name
4. + status → add Finished/DNF info
5. + qualifying → add qualifying position
6. + driver_standings → driver championship position before race
7. + constructor_standings → team championship position before race
8. + pit_stops → how many times driver pitted

```python
# Step 1: results + races
# Adding year, round and circuit info to results
df = results.merge(
    races[['raceId', 'year', 'round', 'circuitId', 'name']], 
    on='raceId', how='left'
)
print(f"Step 1 done: {df.shape}")

# Step 2: Add driver information
# Adding driver name and nationality
df = df.merge(
    drivers[['driverId', 'code', 'forename', 'surname', 'nationality']], 
    on='driverId', how='left'
)
print(f"Step 2 done: {df.shape}")

# Step 3: Add team name
# Replacing constructorId with actual team name
df = df.merge(
    constructors[['constructorId', 'name']], 
    on='constructorId', how='left',
    suffixes=('_circuit', '_team')
)
print(f"Step 3 done: {df.shape}")

# Step 4: Add race status
# Adding Finished / DNF / Accident etc.
df = df.merge(
    status[['statusId', 'status']], 
    on='statusId', how='left'
)
print(f"Step 4 done: {df.shape}")

# Step 5: Add qualifying position
# Where was the driver in qualifying before the race
quali_clean = qualifying[['raceId', 'driverId', 'position']].rename(
    columns={'position': 'quali_position'}
)
df = df.merge(quali_clean, on=['raceId', 'driverId'], how='left')
print(f"Step 5 done: {df.shape}")

# Step 6: Add driver championship standings
# Driver's points and position in championship before this race
ds_clean = driver_standings[['raceId', 'driverId', 'points', 'position']].rename(
    columns={'points': 'driver_pts_before', 
             'position': 'driver_standing_before'}
)
df = df.merge(ds_clean, on=['raceId', 'driverId'], how='left')
print(f"Step 6 done: {df.shape}")

# Step 7: Add constructor championship standings
# Team's points and position in championship before this race
cs_clean = constructor_standings[['raceId', 'constructorId', 'points', 'position']].rename(
    columns={'points': 'team_pts_before', 
             'position': 'team_standing_before'}
)
df = df.merge(cs_clean, on=['raceId', 'constructorId'], how='left')
print(f"Step 7 done: {df.shape}")

# Step 8: Add pit stop count
# How many times each driver pitted in the race
pit_count = pit_stops.groupby(
    ['raceId', 'driverId']
).size().reset_index(name='pit_stop_count')

df = df.merge(pit_count, on=['raceId', 'driverId'], how='left')
print(f"Step 8 done: {df.shape}")

print("\n Main dataframe ready!")
print(f"Total rows    : {df.shape[0]}")
print(f"Total columns : {df.shape[1]}")
print(f"\nAll Columns:\n{df.columns.tolist()}")
```

# Step 4: Checking Missing Values (Null Analysis)

Before building our model, we need to understand:
- Which columns have missing values?
- How many values are missing?
- Why are they missing?
- What should we do about them?

```python
# Total nulls in each column
null_counts = df.isnull().sum()
null_percent = (df.isnull().sum() / len(df) * 100).round(2)

null_df = pd.DataFrame({
    'Missing Count': null_counts,
    'Missing %': null_percent
}).sort_values('Missing %', ascending=False)

# Only show columns that have nulls
null_df = null_df[null_df['Missing Count'] > 0]

print("=== Columns with Missing Values ===")
print(null_df)
```

#  Step 6: Handling Missing Values

## Strategy for each column:

| Column | Missing % | Why? | Solution |
|--------|-----------|------|----------|
| `pit_stop_count` | 79% | Data only from 2011 | Fill with 0 for old races, or filter year >= 2011 |
| `quali_position` | 61% | Old seasons had no qualifying | Fill with `grid` position |
| `team_pts_before` | 7% | First race of season | Fill with 0 |
| `driver_pts_before` | 2% | First race of season | Fill with 0 |
| `team_standing_before` | 7% | First race of season | Fill with 0 |
| `driver_standing_before` | 2% | First race of season | Fill with 0 |

```python
df = df[df['year']>= 2000].copy()
print(f"After filtering year : {df.shape}")

# Strategy 2 — quali_position NaN →  fill with grid position
df['quali_position'] = df['quali_position'].fillna(df['grid'])
print(f"quali_position nulls : {df['quali_position'].isnull().sum()}")

# Strategy 3 — points/standings NaN → fill with 0
df['driver_pts_before'] = df['driver_pts_before'].fillna(0)
df['driver_standing_before'] = df['driver_standing_before'].fillna(0)
df['team_pts_before']        = df['team_pts_before'].fillna(0)
df['team_standing_before']   = df['team_standing_before'].fillna(0)
print("points/standings nulls filled with 0")
```

### Correlation: 0.039 (Very Weak)

### Key Points:

Pit stop count barely affects finish position.
This tells us pit stops alone don't determine result —
it is the TIMING of stops that matters, not the count.

**3. More Stops = Slightly Worse Finish**
Drivers with 5+ stops average P13+ finish —
these are usually drivers who had problems,
punctures, or safety car strategy gambles.

**4. 1 vs 2 Stops — Almost Same Result (P10.2)**
One-stop and two-stop strategies produce identical
average finish — strategy choice depends on
tyre degradation and circuit characteristics.

**5. 7 Stops = P13.7 — Worst Average**
Extreme pit stop counts indicate race incidents —
punctures, damage, or desperate strategy changes.


## 6.8 Year Wise Trends

```python
# Graph 1 — Grid vs Finish correlation by year
corr_by_year = df.groupby('year').apply(
    lambda x: x['grid'].corr(x['positionOrder'])
).reset_index()
corr_by_year.columns = ['year', 'correlation']

fig1 = go.Figure()

fig1.add_trace(go.Scatter(
    x=corr_by_year['year'],
    y=corr_by_year['correlation'],
    mode='lines+markers',
    marker=dict(
        color=corr_by_year['correlation'],
        colorscale='RdYlGn',
        size=8,
        showscale=True,
        colorbar=dict(title='Correlation')
    ),
    line=dict(color='#00D2BE', width=2),
    fill='tozeroy',
    fillcolor='rgba(0, 210, 190, 0.1)'
))

fig1.update_layout(
    title='Grid vs Finish Correlation by Year<br><sup>Higher = Grid position more predictive that year</sup>',
    paper_bgcolor='#1a1a2e',
    plot_bgcolor='#1a1a2e',
    font=dict(color='white'),
    title_font=dict(color='#00D2BE', size=20),
    xaxis_title='Year',
    yaxis_title='Correlation',
    xaxis=dict(gridcolor='#333333'),
    yaxis=dict(gridcolor='#333333'),
)
fig1.show()

# Graph 2 — Average positions gained/lost per year
df['positions_gained'] = df['grid'] - df['positionOrder']
# Positive = gained positions, Negative = lost positions

yearly_gains = df.groupby('year')['positions_gained'].mean().reset_index()

fig2 = go.Figure(go.Bar(
    x=yearly_gains['year'],
    y=yearly_gains['positions_gained'],
    marker=dict(
        color=yearly_gains['positions_gained'],
        colorscale='RdYlGn',
        showscale=True,
        colorbar=dict(title='Avg Positions')
    ),
    text=[f'{v:.2f}' for v in yearly_gains['positions_gained']],
    textposition='outside'
))

fig2.update_layout(
    title='Average Positions Gained/Lost per Year<br><sup>Positive = drivers gained positions on average</sup>',
    paper_bgcolor='#1a1a2e',
    plot_bgcolor='#1a1a2e',
    font=dict(color='white'),
    title_font=dict(color='#00D2BE', size=20),
    xaxis_title='Year',
    yaxis_title='Avg Positions Gained',
    xaxis=dict(gridcolor='#333333'),
    yaxis=dict(gridcolor='#333333', zeroline=True,
               zerolinecolor='white', zerolinewidth=1),
    bargap=0.1
)
fig2.show()

# Graph 3 — Total races per year
races_per_year = df.groupby('year')['raceId'].nunique().reset_index()
races_per_year.columns = ['year', 'total_races']

fig3 = go.Figure(go.Scatter(
    x=races_per_year['year'],
    y=races_per_year['total_races'],
    mode='lines+markers',
    fill='tozeroy',
    fillcolor='rgba(225, 6, 0, 0.1)',
    line=dict(color='#e10600', width=2),
    marker=dict(color='#e10600', size=6)
))

fig3.update_layout(
    title='Total Races per Season (1950-2023)<br><sup>F1 calendar has grown significantly</sup>',
    paper_bgcolor='#1a1a2e',
    plot_bgcolor='#1a1a2e',
    font=dict(color='white'),
    title_font=dict(color='#00D2BE', size=20),
    xaxis_title='Year',
    yaxis_title='Number of Races',
    xaxis=dict(gridcolor='#333333'),
    yaxis=dict(gridcolor='#333333'),
)
fig3.show()

print("=== Recent Years Correlation ===")
print(corr_by_year.tail(10).to_string(index=False))
```
